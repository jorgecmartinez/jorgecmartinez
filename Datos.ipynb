```python
# Balance total de energía
lenght=len(Analisis["EHE(kWh)"])
balance=[]
A=[]
A_limit=1000
A_inicial=1000
B=[]
Venta=[]
Compra=[]
#beneficio=[] #Beneficios del agregador
# lenght2=len(Residencial1["EHE"])
for i in range(lenght):
    balance.append([])
    for j in range(8):
        balance[i].append(0)
i=0
# print(pcr[1],pcg[1])

while i<lenght:
#     EHE_t.append(Residencial1.iloc[i]['EHE']) #Energía excedentaria horaria total
#     EHR_t.append(Residencial1.iloc[i]['EHR']) #Energía a comprar a red horaria total
    B.append(Analisis.iloc[i]["EHE(kWh)"]-Analisis.iloc[i]["EHR(kWh)"])
    # PCR.append(Analisis.iloc[i]["PCR(€/kWh)"])
    # PCG.append(Analisis.iloc[i]["PCG(€/kWh)"])
    if(Analisis.iloc[i]["PCR(€/kWh)"]>Analisis.iloc[i]["PCG(€/kWh)"]):
        if B[i]>=0:
            if i==0:
                if A_inicial+B[i]>A_limit:
                    A.append(A_limit)
                    Venta.append(A_inicial+B[i]-A_limit)
                    Compra.append(0)
                else:
                    A.append(A_inicial+B[i])
                    Compra.append(0)
                    Venta.append(0)
            else:
                if A[i-1]+B[i]>A_limit:
                    A.append(A_limit)
                    Venta.append(A[i-1]+B[i]-A_limit)
                    Compra.append(0)
                else:
                    A.append(A[i-1]+B[i])
                    Compra.append(0)
                    Venta.append(A[i-1]+B[i]-A_limit)      
        else:
            if i==0:
                if (A_inicial>0):
                    if (A_inicial+B[i])<0:
                        A.append(0)
                        Compra.append(-B[i]-A_inicial)
                        Venta.append(0)
                    else:
                        A.append(A_inicial+B[i])
                        Compra.append(0)
                        Venta.append(0)
                else:
                    A.append(0)
                    Compra.append(-B[i])
                    Venta.append(0)
            else:
                if (A[i-1]>0):
                    if (A[i-1]+B[i])<0:
                        A.append(0)
                        Compra.append(-B[i]-A[i-1])
                        Venta.append(0)
                    else:
                        A.append(A[i-1]+B[i])
                        Compra.append(0)
                        Venta.append(0)
                else:
                    A.append(0)
                    Compra.append(-B[i])
                    Venta.append(0)
       
# Subrutina, si PCR<PCG, comprar hasta llenar existencias de almacenamiento
    else:
        # Primer caso
        if i==0: 
            #No tengo espacio de almacenamiento
            if A_inicial==A_limit:
                #Tengo excedente   
                if B[i]>0: 
                    A.append(A_limit)
                    Venta.append(B[i])
                    Compra.append(0)
                
                #No tengo excedente
                else:  
                    A.append(A_limit)
                    Compra.append(-B[i])
                    Venta.append(0)
            #Tengo espacio de almacenamiento
            else:
                #Tengo excedente                       
                if B[i]>0: 
                    A.append(A_limit)
                    Venta.append(0)
                    Compra.append(A_limit-A_inicial-B[i])
                
                #No tengo excedente
                else: 
                    A.append(A_limit)
                    Compra.append(A_limit-A_inicial-B[i])
                    Venta.append(0)
        
        # Resto de casos          
        else:
            #No tengo espacio de almacenamiento
            if A[i-1]==A_limit:
                #Tengo excedente   
                if B[i]>0: 
                    A.append(A_limit)
                    Venta.append(B[i])
                    Compra.append(0)
                
                #No tengo excedente
                else:  
                    A.append(A_limit)
                    Compra.append(-B[i])
                    Venta.append(0)
            #Tengo espacio de almacenamiento
            else:
                #Tengo excedente                       
                if B[i]>0: 
                    A.append(A_limit)
                    Venta.append(0)
                    Compra.append(A_limit-A[i-1]-B[i])
                
                #No tengo excedente
                else: 
                    A.append(A_limit)
                    Compra.append(A_limit-A[i-1]-B[i])
                    Venta.append(0)
# # Subrutina, si PCR>PCG, aprovecho al máximo EHE y el restante lo compro    
#     # else:
    i+=1
i=0
# print(A, Compra, Venta)
while i<lenght:
     balance[i][0]=Analisis.iloc[i]["EHE(kWh)"]
     balance[i][1]=Analisis.iloc[i]["EHR(kWh)"]
     balance[i][2]=B[i]
     balance[i][3]=Analisis.iloc[i]["PCR(€/kWh)"]
     balance[i][4]=Analisis.iloc[i]["PCG(€/kWh)"]
     balance[i][5]=A[i]
     balance[i][6]=round(Compra[i]*Analisis.iloc[i]["PCR(€/kWh)"],3)
     balance[i][7]=round(Venta[i]*Analisis.iloc[i]["PCG(€/kWh)"],3)
     i+=1
Resumen=pd.DataFrame(balance,columns=["EHE(kWh)", "EHR(kWh)", "B(kWh)","PCR(€/kWh)","PCG(€/kWh)", "A(kWh)", "Comprar(€)", "Vender(€)"])
print("Almacenamiento inicial:",A_inicial,"\n")
Resumen.head(5)
# print("Almacenamiento inicial:",A_inicial,"\n", Resumen)
```
